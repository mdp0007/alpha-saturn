# Google News

Uses `googlenewspy` python module to search keywords against the Google News page. Bare bones functionality at the moment, will need to be expanded and/or forked.

```
from googlenewspy import Search
from datetime import datetime
import hashlib

search = Search(host_language="en-US", geolocation="US")
text = "intitle:\"to acquire\""
results = search.news(text)

sortedResults = sorted(
    results,
    key=lambda x: x['date'], reverse=True
)

for result in sortedResults:
    hash_object = hashlib.sha1(result["title"].encode())
    checksum = hash_object.hexdigest()
    print(result["date"])
    print(result["title"])
    print("Checksum: " + checksum)
    print("URL: " + result["url"])
    print("-------")
```
