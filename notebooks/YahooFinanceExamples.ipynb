# Yahoo Finance Examples

This example shows plotting closing prices against an array of tickers over a 30 day preiod.

```
!pip install yfinance
```

```
import yfinance as yf
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter


tickers = ["SUPN", "UNH", "PTLA", "KBH"]

for tick in tickers:
    msft = yf.Ticker(tick)
    msft.history(period="30d")["Close"].plot()
    print("Ticker: " + tick)
    plt.show()
```
